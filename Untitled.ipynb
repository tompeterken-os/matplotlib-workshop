```python
import geopandas as gpd
from shapely.geometry import Polygon
import numpy as np
import matplotlib.pyplot as plt

# Get the world
whole_world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Define some sample bins in longitude
longitude_bin_bounds = np.linspace(-180,180, num=300)
longitude_bin_centres = [0.5*(lo+hi) for lo, hi in zip(longitude_bin_bounds[:-1], longitude_bin_bounds[1:])]
latitude_bin_bounds = np.linspace(-90,90, num=300)
latitude_bin_centres = [0.5*(lo+hi) for lo, hi in zip(latitude_bin_bounds[:-1], latitude_bin_bounds[1:])]

# For each bin we'll need to create a rectangular polygon to intersect
def make_longbin_rect(low, hi):
    return Polygon(((low, -90), (low, 90), (hi, 90), (hi, -90), (low, -90)))
def make_latbin_rect(low, hi):
    return Polygon(((-180, low), (180, low), (180, hi), (-180, hi), (-180, low)))

f, axs = plt.subplots(figsize=(10,8), nrows=2, ncols=2,
                      gridspec_kw={'hspace':0, 'wspace':0, 'height_ratios':[1,2], 'width_ratios':[3,1]},
                      sharex='col', sharey='row')
axs[0,1].axis('off')

for continent, colour in zip(
    ['Oceania', 'Africa', 'North America', 'Asia', 'South America', 'Europe', 'Seven seas (open ocean)', 'Antarctica'],
    ['royalblue', 'red', 'orange', 'green', 'purple', 'grey', 'darkblue', 'lightblue']):
    
    world = whole_world[whole_world['continent']==continent]
    
    long_areas = []

    for bin_num, (low, hi) in enumerate(zip(longitude_bin_bounds[:-1], longitude_bin_bounds[1:])):

        # For each bin, create the polygon rectangle
        bin_poly = make_longbin_rect(low, hi)

        # Find all the bits of the world map within that bin
        world_slice = world.intersection(bin_poly)

        # Find the total land area (requires reprojection to equal-area CRS)
        area = world_slice.to_crs('epsg:3410').area.sum()
        long_areas.append(area)
        
    lat_areas = []

    for bin_num, (low, hi) in enumerate(zip(latitude_bin_bounds[:-1], latitude_bin_bounds[1:])):

        # For each bin, create the polygon rectangle
        bin_poly = make_latbin_rect(low, hi)

        # Find all the bits of the world map within that bin
        world_slice = world.intersection(bin_poly)

        # Find the total land area (requires reprojection to equal-area CRS)
        area = world_slice.to_crs('epsg:3410').area.sum()
        lat_areas.append(area)   
        
        
    # Create a plot
    world.plot(ax=axs[1,0], facecolor=colour)
    axs[1,0].axis('auto')
    axs[0,0].plot(longitude_bin_centres, long_areas, c=colour, label=continent)
    axs[1,1].plot(lat_areas, latitude_bin_centres, c=colour)
    axs[1,0].set_xlabel('Longitude (degrees)')
    axs[1,0].set_ylabel('Latitude (degrees)')
    axs[0,0].set_ylabel('Land area')
    axs[1,1].set_xlabel('Land area')
    
axs[0,0].legend(loc='lower left', bbox_to_anchor=(1.01, 0.01))
[ax.grid('on') for ax in axs.flatten()]


# Find min value
print('MINIMUM AT '+str(np.round(longitude_bin_centres[np.argmin(land_areas)], decimals=1))+' DEGREES')
```
